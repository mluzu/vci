```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

def thresholded(center, pixels):
    out = []
    for p in pixels:
        if p >= center:
            out.append(1)
        else:
            out.append(0)
    return out

def get_pixel_else(m, idx, idy, default=0):
    try:
        return m[idx,idy]
    except IndexError:
        return default


def lbp(file):
    img = cv2.imread(file, 0)
    transformed_img = cv2.imread(file, 0)
    for x in range(0, len(img)):
        for y in range(0, len(img[0])):
            center        = img[x,y]
            top_left      = get_pixel_else(img, x-1, y-1)
            top_up        = get_pixel_else(img, x, y-1)
            top_right     = get_pixel_else(img, x+1, y-1)
            right         = get_pixel_else(img, x+1, y )
            left          = get_pixel_else(img, x-1, y )
            bottom_left   = get_pixel_else(img, x-1, y+1)
            bottom_right  = get_pixel_else(img, x+1, y+1)
            bottom_down   = get_pixel_else(img, x,   y+1 )

            values = thresholded(center, [top_left, top_up, top_right, right, bottom_right,
                                          bottom_down, bottom_left, left])

            weights = [1, 2, 4, 8, 16, 32, 64, 128]
            res = 0
            for a in range(0, len(values)):
                res += weights[a] * values[a]

            transformed_img.itemset((x,y), res)

    return img, transformed_img

def plot(file):
    img, transformed_img = lbp(file)

    hist,bins = np.histogram(img.flatten(),256,[0,256])

    cdf = hist.cumsum()
    cdf_normalized = cdf * hist.max()/ cdf.max()
    
    plt.subplot(1, 2, 1)
    plt.plot(cdf_normalized, color = 'b')
    plt.hist(transformed_img.flatten(),256,[0,256], color = 'r')
    plt.xlim([0,256])
    plt.legend(('cdf','histograma'), loc = 'upper left')
    
    plt.subplot(1, 2, 2)
    plt.imshow(transformed_img, cmap = 'gray')
    
    plt.show()
```

```python
imagenes = ['ladrillos1.png', 'ladrillos2.png', 'oveja1.png', 'oveja2.png', 'piedras1.png', 
            'piedras2.png', 'cielo.png', 'rbt_scn1_focus.png', 'rbt_scn3_focus.png', 'rbt_scn4_focus.png',
            'rbt_scn2.png', 'rbt_scn3.png', 'rbt_scn4.png']

for file in imagenes:
    plot(file)
```
