```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
coord_crom_1 = cv.imread('CoordCrom_1.png', cv.IMREAD_COLOR)


def scale_constrast(img, s):
    img1HSV = cv.cvtColor(coord_crom_1, cv.COLOR_BGR2HSV)
    res = np.tensordot(img1HSV, np.array([[[0, 0, s], [0,0,s], [0,0,s]],
                               [[0, 0, s], [0,0,s], [0,0,s]],
                               [[0, 0, s], [0,0,s], [0,0,s]]]))
    return res

result = scale_constrast(coord_crom_1, 5)
plt.imshow(result)
plt.show()
```

```python
img_wp_r = cv.imread('WP_R.png')
img_wp_g = cv.imread('WP_G.png')
img_wp_b = cv.imread('WP_B.png')


def white_patch(img):
    imgRGB = cv.cvtColor(img, cv.IMREAD_COLOR)
    pixels = np.vstack(imgRGB)
    rgbmax = np.max(pixels, axis=0)
    rgbnorm = np.array([255, 255, 255]) / rgbmax
    norm_img = img * rgbnorm
    return norm_img.astype(int)
  
    

result = white_patch(img_wp_r)
plt.imshow(result)
plt.show()

result = white_patch(img_wp_g)
plt.imshow(result)
plt.show()

result = white_patch(img_wp_b)
plt.imshow(result)
plt.show()


result = white_patch(cv.imread('test_red.png'))
plt.imshow(result)
plt.show()

result = white_patch(cv.imread('test_green.png'))
plt.imshow(result)
plt.show()

result = white_patch(cv.imread('test_blue.png'))
plt.imshow(result)
plt.show()
```
